```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (sepal and petal measurements)
y = iris.target  # True labels (species of iris)
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0)

kmeans.fit(X)

# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Plot the clustered data
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', label='Clustered Data')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')  # Mark centroids
plt.title("K-means Clustering on Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```
