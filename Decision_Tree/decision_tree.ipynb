```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False, parser='auto')
import matplotlib.pyplot as plt
```

```python
x,y=mnist.data,mnist.target
x.shape
```

```python
x
```

```python
y=y.astype(np.uint8)
y
```

```python
random_index = np.random.randint(0, x.shape[0])
plt.imshow(x[random_index].reshape(28,28), cmap='gray')
plt.show()
```

```python
X_train, X_test, y_train, y_test = x[:49000], x[49000:], y[:49000], y[49000:]
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
y_train5=(y_train==5)
y_test5=(y_test==5)
y_test5
```

```python
clf = DecisionTreeClassifier(max_depth=10, random_state=42)
```

```python

clf.fit(X_train, y_train5)
```

```python
y_pred = clf.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test5, y_pred)
print("Test Accuracy:", accuracy*100)

```

```python
clf = DecisionTreeClassifier(max_depth=15, random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```

```python
fig, axes = plt.subplots(10, 10, figsize=(10, 10))
axes = axes.ravel()  

for i in range(100):
    axes[i].imshow(X_test[i].reshape(28, 28), cmap='gray')
    axes[i].set_title(f"Pred: {y_pred[i]}")
    axes[i].axis('off')  

plt.subplots_adjust(hspace=0.7)
plt.show()
```
